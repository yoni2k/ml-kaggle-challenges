```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
train = pd.read_csv('input/train.csv')
train.head()
```

```python
test = pd.read_csv('input/test.csv')
test.head()
```

```python
def extract_lastname(full_name):
    return full_name.split(',')[0]
```

```python
train['Last name'] = train['Name'].apply(extract_lastname)
train['Last name'].value_counts().head()
```

```python
train['Cabin'].fillna(value='', inplace=True)
train['Deck'] = 'None'
train['Deck'] = train[train['Cabin'].isnull()==False]['Cabin'].apply(lambda cabin: cabin[0] if cabin != '' else 'None')
```

```python
last_names_num_tickets = train.groupby(by=['Last name'])['Ticket'].nunique()
last_names_num_tickets[last_names_num_tickets > 1]
```

```python
last_names_numerous_tickets = last_names_num_tickets[last_names_num_tickets > 1].index
last_names_numerous_tickets
```

```python
train_mult_tick = train[train['Last name'].map(lambda x: x in list(last_names_numerous_tickets))]
train_mult_tick
```

```python
mult_num_pclass = train_mult_tick.groupby(by='Last name')['Pclass'].nunique()
same_last_diff_class = list(mult_num_pclass[mult_num_pclass > 1].index)
print(f'Num same last names different class: {len(same_last_diff_class)}, same_last_diff_class: {same_last_diff_class}') 
```

```python
mult_num_deck = train_mult_tick.groupby(by='Last name')['Deck'].nunique()
same_last_diff_deck = list(mult_num_deck[mult_num_deck > 1].index)
print(f'Num same last names different deck: {len(same_last_diff_deck)}, same_last_diff_class: {same_last_diff_deck}') 
```

```python
mult_num_embarked = train_mult_tick.groupby(by='Last name')['Embarked'].nunique()
same_last_diff_embarked = list(mult_num_embarked[mult_num_embarked > 1].index)
print(f'Num same last names different embarked: {len(same_last_diff_embarked)}, same_last_diff_class: {same_last_diff_embarked}') 
```

```python
same_last_something_different = set(same_last_diff_class) | set(same_last_diff_deck) | set(same_last_diff_embarked)
print(f'Num same last names different SOMETHING: {len(same_last_something_different)}, same_last_diff_class: {same_last_something_different}') 
```

```python
same_last_all_different = set(same_last_diff_class) & set(same_last_diff_deck) & set(same_last_diff_embarked)
print(f'Num same last names different EVERYTHING: {len(same_last_all_different)}, same_last_diff_class: {same_last_all_different}') 
```

```python
set_12 = set(same_last_diff_class) & set(same_last_diff_deck)
set_23 = set(same_last_diff_deck) & set(same_last_diff_embarked)
set_13 = set(same_last_diff_class) & set(same_last_diff_embarked)
set_2_diff = set_12 | set_23 | set_13
print(f'Num same last names 2 different: {len(set_2_diff)}, same_last_diff_class: {set_2_diff}') 
```

```python
same_last_something_different - set_2_diff
```

## Bottom line: 
There is an issue today with relying on last name for the feature of `% survival rate` of the family, since there are common last names that are probably not the same family.

We see that 8 families appear in test group that also different at least by 2 out of 3 parameters of `Pclass`, `Embarked`, `Deck`.  There is a good chance that for these familities we have a lot of bad predictions since we are relying on `% survival rate`, but either it's not accurate already in training set, or it's not the same family in test set.  

Best (and probably would improve accuracy of the model) would be to use the information above not to combine into the same family last names that don't share the same `Deck` / `Embarked` port / `Pclass`, or differ in 2 out of 3 parameters.
This would probably make the feature `% survival rate` of the family available for less families, but stronger.

**However**, since the feature of `% survival rate` is very complicated already, decided not to do it currently.
